```python
import pandas as pd
dataset = pd.read_csv("C:/Users/HP/Downloads/iris.csv")

```

```python
dataset
```

```python
dataset['species'].unique()
```

```python
dataset['species'].replace({'setosa' : '1',  'versicolor': '2', 'virginica':'3'}, inplace = True)
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset[['sepal_length' , 'sepal_width', 'petal_length', 'petal_width']],dataset['species'],test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test, y_test)
```
